## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv(r'C:\Users\yuka6\OneDrive\デスクトップ\PG\BasicPython\data\hist_data.csv')
fig, ax = plt.subplots()
plt.hist(df['aデータ'], alpha = 0.5, bins = 30, label = 'aデータ', color ='orange')
plt.hist(df['bデータ'], alpha = 0.5, bins = 30, label = 'bデータ', color = 'lightgreen')
plt.legend(loc = 'upper right', fontsize = 13)
ax.set_xlabel("温度(℃)")
ax.set_ylabel("度数")
ax.set_title("aデータとbデータの温度に関する度数分布")
print("get_xlabel()", ax.get_xlabel())
print("get_ylabel()", ax.get_ylabel())
print("get_title", ax.get_title())
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv(r"C:\Users\yuka6\OneDrive\デスクトップ\PG\BasicPython\data\scatter_data.csv")
df_0 = df[df['class'] == 0]
df_1 = df[df['class'] == 1]
df_2 = df[df['class'] == 2]
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.grid(True)
plt.legend(loc = 'upper right')
plt.scatter(df_0['length (cm)'], df_0['width (cm)'], c = 'r', label = 'class0')
plt.scatter(df_1['length (cm)'], df_1['width (cm)'], c = 'b', label = 'class1')
plt.scatter(df_2['length (cm)'], df_2['width (cm)'], c = 'g', label = 'class2')
plt.legend(loc = 'upper right')
plt.show()
```

```python
df = pd.read_csv(r"C:\Users\yuka6\OneDrive\デスクトップ\PG\BasicPython\data\line_data.csv")
x = df['期間']
y = df['件数']
plt.grid()
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.plot(x, y, marker = 'o')
plt.xticks(rotation = 30)
```
